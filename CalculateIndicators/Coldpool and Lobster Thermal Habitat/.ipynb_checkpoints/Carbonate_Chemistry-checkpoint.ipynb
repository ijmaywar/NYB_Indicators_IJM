```python
import pandas as pn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
import xarray as xr
```

```python
# This cell just defines the function that extracts the transect 1 data from a csv file.  The output is just a table/dataframe (just like in R!)
# with only the values for transect 1
def transect_1(csv_file):
    df = pn.read_csv(csv_file)
    
    # turn the numbered stations into floats
    number_station = np.empty(len(df.Station))
    for i in range(len(df)):
        if type(df.Station[i]) == str and (len(df.Station[i])>4):
            number_station[i] = np.nan
        else:
            number_station[i] = float(df.Station[i])
    
    # add a column for this new float
    df['station_number'] = number_station
    
    if np.any(np.unique(np.floor(df.station_number)) == 1):
        transect1 = df[np.floor(df.station_number) == 1]
    else:
        print('No transect 1')
            
    distance1 = np.empty(len(np.where((np.floor(np.unique(df.station_number))==1))[0]))
    distance1[:] = np.nan

    R = 6373.0 #radius of the Earth
    D = 0
    stations1 = np.unique(transect1.station_number)
    
    for i in range(len(stations1)-1):
        station_subset = transect1[transect1.station_number == stations1[i]]
        station_subset_next = transect1[transect1.station_number == stations1[i+1]]
    
        lat1 = np.radians(np.unique(station_subset.Lat)[0]) #coordinates
        lon1 = np.radians(np.unique(station_subset.Lon)[0])

        lat2 = np.radians(np.unique(station_subset_next.Lat)[0])
        lon2 = np.radians(np.unique(station_subset_next.Lon)[0])

        dlon = lon2 - lon1 #change in coordinates
        dlat = lat2 - lat1

        a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2 #Haversine formula
        c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))

        distance1[i] = (R * c)
        
    transect1['along_track_distance'] = np.nan
    transect1.along_track_distance[transect1.station_number == stations1[0]] = 0
    for i in range(len(stations1)-1):
        transect1.along_track_distance[transect1.station_number == stations1[i+1]] = np.sum(distance1[:i+1])

    return (transect1)   
```

```python
# This cell just defines the function that extracts the transect 6 data from a csv file.  The output is just a table/dataframe (just like in R!)
# with only the values for transect 6
def transect_6(csv_file):
    df = pn.read_csv(csv_file)
    
    # turn the numbered stations into floats
    number_station = np.empty(len(df.Station))
    for i in range(len(df)):
        if type(df.Station[i]) == str and (len(df.Station[i])>4):
            number_station[i] = np.nan
        else:
            number_station[i] = float(df.Station[i])
    
    # add a column for this new float
    df['station_number'] = number_station
    
    if np.any(np.unique(np.floor(df.station_number)) ==6):
        transect6 = df[np.floor(df.station_number) == 6]
    else:
        print('No transect 6')
            
    
    distance6 = np.empty(len(np.where((np.floor(np.unique(df.station_number))==6))[0]))
    distance6[:] = np.nan

    R = 6373.0 #radius of the Earth
    D = 0

    stations6 = np.unique(transect6.station_number)
    
    for i in range(len(stations6)-1):
        station_subset = transect6[transect6.station_number == stations6[i]]
        station_subset_next = transect6[transect6.station_number == stations6[i+1]]
    
        lat1 = np.radians(np.unique(station_subset.Lat)[0]) #coordinates
        lon1 = np.radians(np.unique(station_subset.Lon)[0])

        lat2 = np.radians(np.unique(station_subset_next.Lat)[0])
        lon2 = np.radians(np.unique(station_subset_next.Lon)[0])

        dlon = lon2 - lon1 #change in coordinates
        dlat = lat2 - lat1

        a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2 #Haversine formula
        c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))

        distance6[i] = (R * c)

        
    transect6['along_track_distance'] = np.nan
    transect6.along_track_distance[transect6.station_number == stations6[0]] = 0
    for i in range(len(stations6)-1):
        transect6.along_track_distance[transect6.station_number == stations6[i+1]] = np.sum(distance6[:i+1])
    return (transect6)   
```

```python
transect_1_lon = [-71.9,-71.89916667,-71.81083333,-71.72033333,-71.63483333,-71.54466667,-71.46083333]
transect_1_lat = [41.001,40.96066667,40.75866667,40.86946667,40.47173333,40.0445,39.80066667]
topo = xr.open_dataset("/Users/nyelab/Downloads/ETOPO1_Bed_g_gmt4(1).grd")
transect_1_seafloor_depth = np.empty(7)
transect_1_seafloor_depth[:] = np.nan

for i in range(len(transect_1_seafloor_depth)):

    lonvalue = transect_1_lon[i]
    latvalue = transect_1_lat[i]

    absolute_val_array_lon = np.abs(topo.x - lonvalue)
    absolute_val_array_lat = np.abs(topo.y - latvalue)

    smallest_difference_index_lon = absolute_val_array_lon.argmin()
    smallest_difference_index_lat = absolute_val_array_lat.argmin()

    transect_1_seafloor_depth[i] = topo.z[smallest_difference_index_lat,smallest_difference_index_lon]

distance1 = np.empty(6)
distance1[:] = np.nan

R = 6373.0 #radius of the Earth
D = 0

    
for i in range(6):
    
    lat1 = np.radians(transect_1_lat[i]) #coordinates
    lon1 = np.radians(transect_1_lon[i])

    lat2 = np.radians(transect_1_lat[i+1])
    lon2 = np.radians(transect_1_lon[i+1])

    dlon = lon2 - lon1 #change in coordinates
    dlat = lat2 - lat1

    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2 #Haversine formula
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1 - a))

    distance1[i] = (R * c)
i=0
distance_t1 = [0,np.sum(distance1[i:i+1])*1000,np.sum(distance1[i:i+2])*1000,
              np.sum(distance1[i:i+3])*1000,np.sum(distance1[i:i+4])*1000,
              np.sum(distance1[i:i+5])*1000,np.sum(distance1[i:i+6])*1000]

# linear interpolation of seafloor depth between ctd cast points
y2 = transect_1_seafloor_depth
x2 = distance_t1
xq = np.linspace(0,distance_t1[6], 17)
y_new = np.interp(xq,x2,y2)

depth = y_new
depth[depth<-100] = -100

D = np.diff(xq)
d = np.diff(depth)
area = np.empty(16)

for i in range(16):
    X = D[i]/1000
    Y = depth[i]/1000
    area_rect = X*Y
    Y2 = d[i]/1000
    area_triangle = 0.5 * X*Y2
    area[i] = (area_rect*(-1)) + (area_triangle*(-1))
area

transect1_area = np.sum(area)
```

```python
# This defines a function that allows you to pick a variable from your transect data frame, interpolate the data from the different CTD casts,
# plots the interpolated data on a graph with the contour level of your choosing highlighted and with a title that you pick
# It also spits out the area within that contour level and the mean value of the variable within that contour level

def area_within_contour(transect_data, variable, contour_level, title_of_plot, T1_or_T6):
    # Next establish the seafloor depth
    topo = xr.open_dataset("/Users/nyelab/Downloads/ETOPO1_Bed_g_gmt4(1).grd")
    transect = transect_data #fill in the name of the transect

    #  Find the depth at each ctd cast site
    lons = np.unique(transect['Lon'])
    lats = np.flip(np.unique(transect['Lat']))
    seafloor_depth = np.empty(len(np.unique(transect['Lat'])))
    seafloor_depth[:] = np.nan
    for i in range(len(seafloor_depth)):

        lonvalue = lons[i]
        latvalue = lats[i]

        absolute_val_array_lon = np.abs(topo.x - lonvalue)
        absolute_val_array_lat = np.abs(topo.y - latvalue)

        smallest_difference_index_lon = absolute_val_array_lon.argmin()
        smallest_difference_index_lat = absolute_val_array_lat.argmin()

        seafloor_depth[i] = topo.z[smallest_difference_index_lat,smallest_difference_index_lon]


    print('seafloor depths,', seafloor_depth)
    
    # Preforma a gridded interpolation of whatever variable you are looking at
    y= np.arange(0,110,10)
    x= np.arange(0,240000, 10000)
    X, Y = np.meshgrid(x,y)
    grid_transect = sio.griddata((transect['along_track_distance']*1000, transect['Depth [m]']), transect[variable],(X,Y))
    
    # linear interpolation of seafloor depth between ctd cast points
    y2 = seafloor_depth
    x2 = np.unique(transect['along_track_distance']*1000)
    xq = x
    y_new = np.interp(xq,x2,y2)
    
    # delete all the interpolated stuff below the seafloor depth
    # Find the depth at each ctd cast site
    delete_index = np.empty(len(xq))
    for i in range(len(xq)):

        value = abs(y_new[i])

        absolute_val_array = np.abs(y - value)

        smallest_difference_index = absolute_val_array.argmin()

        delete_index[i] = smallest_difference_index +1
        
    for i in range(len(xq)):
        num = int(delete_index[i])
        grid_transect[num:,i] = np.nan  
        
    # Plot the data
    fig, ax = plt.subplots(figsize = [25,10])
    c = plt.contourf(x,y,grid_transect, levels = 15, cmap = 'rainbow')
    c2 = plt.contour(x,y,grid_transect,levels = [contour_level])
    plt.ylim([100,0])
    plt.colorbar(c)
    plt.clabel(c2)
    plt.scatter(transect['along_track_distance']*1000, transect['Depth [m]'],c = 'k')
    ax.tick_params(labelsize = 16)
    ax.set_ylabel('Depth (m)', fontsize = 16)
    ax.set_xlabel('Along Section Distance (m)', fontsize = 16)
    ax.set_title(title_of_plot, fontsize = 18)
    
    area_within_contour = (len(grid_transect[grid_transect<contour_level])*10*10000)/(1000*1000)
    average = np.nanmean(grid_transect[grid_transect<contour_level])
    
    if T1_or_T6 == 'T1':
        percent_area = (area_within_contour/transect1_area)*100
    if T1_or_T6 == 'T6':
        percent_area = (area_within_contour/transect6_area)*100
        
    print('area within contour is ', area_within_contour, ' km')
    print('percent transect area within contour is ', percent_area, '%')
    print('average value within contour is ', average)
    return(area_within_contour)
```
